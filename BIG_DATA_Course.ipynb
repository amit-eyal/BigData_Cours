```python
# class4)

# הצג לכל לקוח את סכום הפדיון הגדול מחמש מאות
# רק עבור לקוחות מערים שמכילות בשמם את האותיות איי בי וסי
# והפדיון שלהם מכל מוצר בכל הזמנה קטן מעשרים וגדול ממאה

# פדיון זה הכמות כפול המחיר

# מהן הטבלאות - from
# מהן הקשרים - from
# מי הן המפתחות - from
# WHERE
# לפי מה מקבצים - group by
# מה סינון לאחר קיבוץ - having
SELECT c.customer_id, c.company_name, SUM(od.quantity* od.price) AS 'revenue'
FROM customers as c INER JOIN orders as o INER JOIN [order ditails] as od
     ON od.order_id = o.order_id
     ON o.customer_id = c.customer_id
WHERE c.city_name like '[%abc%]' AND od.quantity*od.price NOT BETWEEN 20 AND 100
GROUP BY c.customer_id, c.company_name
HAVING SUM(od.quantity* od.price) > 500
```

```python
# class5)

# הצג לכל רבעון ללא קשר לשנה את סכום עלויות ההובלה הגדולות מחמש מאות
# עבור לקוחות שקוד הלקוח מתחיל באותיות איי בי וסי יש למיין את התוצאות בסדר יורד על פי סכום העלויות
SELECT DATEPART(QUARTER, o.order_date) AS 'q', SUM(o.freight) AS 'f'
FROM orders as o
WHERE o.customer_id LIKE '[%abc%]'
GROUP By DATEPART(QUARTER, o.order_date)
HAVING SUM(o.freight) > 500
ORDER BY o.f DESC

# הצג לכל לקוח את סכום הפדיון הגבוה מאלף חמש מאות
# רק עבור מוצרים שהאיכות מוצר מכילה את המילה קופסאות וכמותם במלאי גדול מחמישים וקטן מ שמונה עשרה
quantityPerUnit, Boxes

SELECT c.customer_id, c.company_name, SUM(od.quantity* od.price)
FROM customers as c INER JOIN orders as o INER JOIN [order ditails] as od INER JOIN products AS p
     ON p.product_id = od.product_id
     ON od.order_id = o.order_id
     ON o.customer_id = c.customer_id
WHERE od.quantityPerUnit LIKE '[%Boxes%]' AND p.UnitInStock NOT BETWEEN 18 AND 50
GROUP By c.customer_id, c.company_name
HAVING SUM(od.quantity* od.UnitPrice) > 1500

# הצג לכל חודש הזמנה ללא קשר לשנה את ממוצע עלות ההובלה וכמות ההזמנות
SELECT MONTH(o.order_date), AVG(o.freight), COUNT(*) AS 'count of orders'
FROM orders as o
GROUP BY MONTH(o.order_date)

# הצג לכל לקוח בכל ההזמנות שלו מהו התאריך הזמנה הראשון והאחרון ומרחק הימים בין התאריכים
SELECT c.customer_id, c.company_name, MAX(o.order_date), MIN(o.order_date), DATEDIFF(DAY, MIN(o.order_date), MAX(o.order_date)
FROM customers AS c INER JOIN orders AS o
     ON o.customer_id = c.customer_id
GROUP BY c.customer_id, c.company_name

```

```python
# class6)

# הצג לכל שם סטודנט את שמות הקורסים שלו
SEKECT s.student_name, c.curse_name
FROM students AS s INER JOIN students_courses AS sc INER JOIN courses AS c
     ON c.curse_id = sc.curse_id
     ON sc.student_id = s.student_id

# הצג את הלקוחות שלא ביצעו הזמנה
SELECT c.customer_name
FROM customers AS c LEFT JOIN orders AS o
     ON o.customer_id = c.customer_id
WHERE o.order_date IS NULL

//

SELECT c.customer_name
FROM customers AS c OUTER JOIN orders AS o
     ON o.customer_id = c.customer_id

# הצג לכל עובד את כלל ההזמנות שהוא טיפל בהם
SELECT e.*, o.*
FROM employees AS e LEFT JOIN orders AS o
     ON o.employee_id = e.employee_id

# הצג ספקים שלא סיפקו מוצרים
SELECT s.* p.*
FROM suppliers AS s LEFT JOIN products AS p
     ON p.supplier_id = s.supplier_id
WHERE p.product_name IS null

//

SELECT s.*, p.*
FROM suppliers AS S OUTER JOIN products AS P
     ON p.suppliers_id = s.products_id

# הצג בסדר יורד דוח ללקוחות וספקים הגרים בברזיל
# יש להציג שם לקוח/ ספק, איש קשר, טלפון, מידע האם לקוח או ספק

SELECT c.customer_name, c.contact_name, c.phone, 'customers' AS INFO
FROM customers AS c
WHERE c.contry LIKE 'BRAZIL'
ONION
s.suppliers_name, s.contact_name, s.phone, 'suppliers' AS INFO
FROM suppliers AS s
WHERE s.contry LIKE 'BRAZIL'
ORDER BY s.supplier_name DESC

# הצג דוח לקוחות והזמנות לפי הפירוט הבא
# לכל לקוח הזמנות פלוס פדיון
# סכום פדיון ופרטים נוספים לכל לקוח
# סכום פדיון כללי בארגון
SELECT c.customer_id, c.company_name, o.order_date, o.freight, od.quantity* od.price AS 'revenue'
FROM customers as c INER JOIN orders as o INER JOIN [order ditails] as od
     ON od.order_id = o.order_id
     ON o.customer_id = c.customer_id
UNION
SELECT c.customer_id, 'TOTAL', max(o.order_date), MIN(o.freight), SUM(od.quantity* od.price) AS 'revenue'
FROM customers as c INER JOIN orders as o INER JOIN [order ditails] as od
     ON od.order_id = o.order_id
     ON o.customer_id = c.customer_id
GROUP BY c.customer_id
UNION
SELECT 'total all', '', '', '', SUM(od.quantity* od.price) AS 'revenue'
FROM [order details] as od
```

```python
# class7)

# הצג לכל לקוח את עלות ההובלה הגדולה מממוצע עליות ההובלה של כלל הארגון
SELECT c.customer_id, c.company_name, o.order_date, o.freight
FROM customers AS c INNER JOIN orders AS o
     ON o.customer_id = c.customer_id
WHERE o.freight >
(SELECT AVG(o.freight)
FROM orders AS o)

# הצג את שמות הלקוחות אשר הזמינו בתאריך בו הזמין לקוח בשם הנרי את הזמנתו האחרונה
SELECT c.customer_id, c.company_name, o.order_date
FROM customers AS c INNER JOIN orders AS o
    ON o.customer_id = c.customer_id
WHERE o.order_date =
      (SELECT MAX(o.order_date)
       FROM customeres AS c orders AS o
            ON o.customer_id = c.customer_id
       WHERE c.company_name LIKE 'hanari carnes')

# מי הם הספקים אשר סיפקו מוצרים אותם רכש לקוח שהקוד שלו הוא אלפקי
SELECT DISTINCT s.company_name, s.supllier_id
FROM suppliers AS s INNER JOIN products AS p
     ON p.supplier_id = s.supplier_id
WHERE p.product_id IN
      (SELECT od.product_id
      FROM customers AS c INNER JOIN orders AS o INNER JOIN [order details] AS od INNER JOIN products AS p
          ON od.order_id = o.order_id
          ON o.customer_id = c.customer_id
      WHERE c.customer_id LIKE 'ALFKI')
```

```python
# class8)

#הצג לכל הזמנה מוצרים רק עבור מוצרים שכמותם במלאי גדולים משלושים וקטנים משמונה
#ועלות הובלה בין חמש ל15 ונשלחו בזמן
SELECT *
FROM orders AS o INNER JOIN [order details] AS od INNER JOIN products AS p
     ON od.product_id = p.product_id
     ON od.order_id = o.order_id
WHERE p.unit_in_stock NOT BETWEEN 8 AND 30
      AND od.freight BETWEEEN 5 AND 15
      AND o.shipped_date <= o.required_date

# הצג לכל לקוח את סכום הפדיון הגדול מחמש מאות רק עבור לקוחות מהעיר פריז
# שהזמינו הזמנות בתאריך שלקוח שהקוד שלו הוא אלפקי הזמין

SELECT c.customer_id, c.company_name, c.city, SUM(od.quantity* od.price)
FROM customers as c INER JOIN orders as o INNER JOIN [order ditails] as od
     ON od.order_id = o.order_id
     ON o.customer_id = c.customer_id
WHERE c.city IN ('FRANCE', 'LONDON') AND o.order_date IN
      (SELECT DISTINCT o.order_date
      FROM orders as o
      WHERE o.customer_id LIKE 'ALFKI')
GROUP BY c.customer_id, c.company_name
HAVING SUM(od.quantity* od.price) > 500

# הצג את זוגות העובדים שגרים באותה עיר
SELECT e1.first_name, e2.first_name, e1.city
FROM employees AS e1 INNER JOIN employees AS e2
     ON e1.employee_id < e2.employee_id
WHERE e1.city = e2.city
```
